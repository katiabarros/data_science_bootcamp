### Creating dummies and checking on it.

```python
reason_columns=pd.get_dummies(df['Reason for Absence'])
reason_columns
```

```python
reason_columns['check'].unique()
```

```python
reason_columns=reason_columns.drop(['check'],axis=1)
reason_columns
```

To avoid multicollinearity, we remove the first column with dummy variables.

```python
reason_columns=pd.get_dummies(df['Reason for Absence'],drop_first=True)
reason_columns
```

### Group the reasons for absence

```python
df.columns.values
```

```python
df=df.drop(['Reason for Absence'],axis=1)
df
```

```python
#Group 1: between 1 and 14, Group 2: 15 and 17, Group 3: 18 and 21, Group 4: 22 and 28
reason_type_1=reason_columns.loc[:,1:14].max(axis=1)
reason_type_2=reason_columns.loc[:,15:17].max(axis=1)
reason_type_3=reason_columns.loc[:,18:21].max(axis=1)
reason_type_4=reason_columns.loc[:,22:].max(axis=1)
```

### Concatenate Column Values

```python
df
```

```python
df=pd.concat([df,reason_type_1,reason_type_2,reason_type_3,reason_type_4],axis=1)
df
```

## Date

```python
type(df_reason_mod['Date'])
```

```python
df_reason_mod['Date']
```

```python
df_reason_mod['Date']=pd.to_datetime(df_reason_mod['Date'],format='%d/%m/%Y')  #converts value to timestamp
df_reason_mod['Date']
```

### Extract the month value

```python
df_reason_mod['Date'][0].month
```

```python
list_months=[]
for i in range (700):
 list_months.append(df_reason_mod['Date'][i].month)
list_months
```

```python
df_reason_mod['Month Value']=list_months
df_reason_mod
```

### Extract the day of the week

```python
#Monday = 0, ..., Sunday = 6
df_reason_mod['Date'][0].weekday()
```

```python
def date_to_weekday(date_value):
 return date_value.weekday()
```

```python
df_reason_mod['Day of the week']=df_reason_mod['Date'].apply(date_to_weekday)
```

```python
df_reason_mod
```

```python
df_reason_mod
```

```python
df_reason_mod=df_reason_mod.drop(['Date'],axis=1)
```
